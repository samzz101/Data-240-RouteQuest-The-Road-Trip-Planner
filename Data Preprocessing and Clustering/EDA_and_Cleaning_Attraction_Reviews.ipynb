```python
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split
```

```python
# Load the dataset into the Surprise library
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df[['user', 'attraction_id', 'rating']], reader)

# Split the data into training and testing sets
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# Build the user-based collaborative filtering model
sim_options = {
    'name': 'cosine',  # Use cosine similarity
    'user_based': True  # Use user-based approach
}
model = KNNBasic(sim_options=sim_options)

# Train the model on the training set
model.fit(trainset)

# Function to get attraction recommendations for a given user
def get_attraction_recommendations(user_id, num_recommendations=5):
    # Get a list of all attraction IDs
    all_attractions = df['attraction_id'].unique()

    # Predict ratings for attractions not rated by the user
    predictions = [model.predict(user_id, attraction_id) for attraction_id in all_attractions]

    # Sort predictions by estimated rating in descending order
    sorted_predictions = sorted(predictions, key=lambda x: x.est, reverse=True)

    # Return the top N recommendations
    top_recommendations = [(prediction.iid, prediction.est) for prediction in sorted_predictions[:num_recommendations]]

    return top_recommendations
```

```python
# Example: Recommend attractions for a user (replace 'user123' with an actual user ID)
user_id_to_recommend = 'user123'
recommendations = get_attraction_recommendations(user_id_to_recommend)

# Display the recommendations
print(f"Recommendations for {user_id_to_recommend}: {recommendations}")
```

```python
import numpy as np
# Identify outliers using the IQR method
Q1 = merged_df['normalized_popularity'].quantile(0.25)
Q3 = merged_df['normalized_popularity'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = (merged_df['normalized_popularity'] < lower_bound) | (merged_df['normalized_popularity'] > upper_bound)

# Calculate the maximum non-outlier value
max_non_outlier = np.max(merged_df.loc[~outliers, 'normalized_popularity'])

# Replace outliers with the maximum non-outlier value
merged_df.loc[outliers, 'normalized_popularity'] = max_non_outlier
```
