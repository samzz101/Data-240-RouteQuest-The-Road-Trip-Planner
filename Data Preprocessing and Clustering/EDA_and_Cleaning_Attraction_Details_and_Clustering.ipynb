```python
#Normalize the price column to 1-5 range

import csv

# Read data from the CSV file
with open('/content/drive/Shareddrives/DATA 240- Data Mining/Cleaned dataset and code/Canada /Final Cleaned/cleaned.csv', mode='r') as infile:
    reader = csv.DictReader(infile)
    data = list(reader)

# Extract prices and find min and max
prices = [float(row['price']) for row in data]
min_price = min(prices)
max_price = max(prices)

# Normalize prices to a 1-5 scale
for row in data:
    normalized_price = 1 + (float(row['price']) - min_price) / (max_price - min_price) * 4
    row['price'] = round(normalized_price)

# Write data back to the CSV file
with open('/content/drive/Shareddrives/DATA 240- Data Mining/Cleaned dataset and code/Canada /Final Cleaned/cleaned2.csv', mode='w', newline='') as outfile:
    writer = csv.DictWriter(outfile, fieldnames=data[0].keys())
    writer.writeheader()
    writer.writerows(data)
```

```python
import csv

# Define the classification function
def classify_sentiment(score):
    if score < 0:
        return 'Bad'
    elif score < 0.35:
        return 'OK'
    elif score < 0.45:
        return 'Good'
    else:
        return 'Very Good'

# Paths to the input and output CSV files
input_file_path = '/content/drive/Shareddrives/DATA 240- Data Mining/Cleaned dataset and code/Canada /Final Cleaned/cleaned2.csv'
output_file_path = '/content/drive/Shareddrives/DATA 240- Data Mining/Cleaned dataset and code/Canada /Final Cleaned/cleaned3.csv'

# Read the data and write to a new file with an additional column for sentiment rating
with open(input_file_path, mode='r', newline='') as infile, \
     open(output_file_path, mode='w', newline='') as outfile:

    reader = csv.DictReader(infile)
    fieldnames = reader.fieldnames + ['sentiment_rating']  # New column for sentiment rating
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)

    writer.writeheader()
    for row in reader:
        sentiment_score = float(row['sentiment_y'])
        row['sentiment_rating'] = classify_sentiment(sentiment_score)
        writer.writerow(row)

print(f"New CSV file with sentiment rating created at: {output_file_path}")
```

```python
import csv

# Define the classification function for normalized popularity
def classify_popularity(score):
    if score < -0.2:
        return 'Not Popular'
    elif -0.2 <= score < 0.2:
        return 'Popular'
    else:
        return 'Very Popular'

# Paths to the input and output CSV files
input_file_path = '/content/drive/Shareddrives/DATA 240- Data Mining/Cleaned dataset and code/Canada /Final Cleaned/cleaned3.csv'  # You will need to update this path
output_file_path = '/content/drive/Shareddrives/DATA 240- Data Mining/Cleaned dataset and code/Canada /Final Cleaned/cleaned4.csv'  # You will need to update this path

# Read the data and write to a new file with an additional column for popularity rating
with open(input_file_path, mode='r', newline='') as infile, \
     open(output_file_path, mode='w', newline='') as outfile:

    reader = csv.DictReader(infile)
    fieldnames = reader.fieldnames + ['popularity_rating']  # New column for popularity rating
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)

    writer.writeheader()
    for row in reader:
        popularity_score = float(row['normalized_popularity'])
        row['popularity_rating'] = classify_popularity(popularity_score)
        writer.writerow(row)

print(f"New CSV file with popularity rating created at: {output_file_path}")
```

```python
import csv

# Function to convert price level to dollar signs
def price_to_dollar_signs(price_level):
    return '$' * int(price_level)  # Multiplies the $ sign by the price level

# Paths to the input and output CSV files
input_file_path = '/content/drive/Shareddrives/DATA 240- Data Mining/Cleaned dataset and code/Canada /Final Cleaned/cleaned4.csv'  # You will need to update this path
output_file_path = '/content/drive/Shareddrives/DATA 240- Data Mining/Cleaned dataset and code/Canada /Final Cleaned/cleaned5.csv'  # You will need to update this path

# Read the data and write to a new file with an additional column for dollar signs
with open(input_file_path, mode='r', newline='', encoding='utf-8') as infile, \
     open(output_file_path, mode='w', newline='', encoding='utf-8') as outfile:

    reader = csv.DictReader(infile)
    fieldnames = reader.fieldnames + ['price_level']  # New column for price level as dollar signs
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)

    writer.writeheader()
    for row in reader:
        price_level = row['price']  # Assuming this is the name of your price column
        row['price_level'] = price_to_dollar_signs(price_level)
        writer.writerow(row)

print(f"New CSV file with price levels created at: {output_file_path}")
```

```python
import csv
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Path to your CSV file
file_path = '/content/drive/Shareddrives/DATA 240- Data Mining/Cleaned dataset and code/Canada /Final Cleaned/full_cleaned_normalized.csv'

# Read the data from the CSV file
with open(file_path, mode='r') as file:
    reader = csv.DictReader(file)
    data = list(reader)

# Extract the relevant features for clustering
# Convert each column to a float and store in a list
prices = [float(item['price']) for item in data]
ratings = [float(item['rating']) for item in data]
latitudes = [float(item['latitude']) for item in data]
longitudes = [float(item['longitude']) for item in data]
sentiment_ys = [float(item['sentiment_y']) for item in data]
popularities = [float(item['popularity']) for item in data]
normalized_popularities = [float(item['normalized_popularity']) for item in data]

# Combine the features into a NumPy array
features = np.column_stack((prices, ratings, latitudes, longitudes, sentiment_ys, popularities, normalized_popularities))

# Select a random subset of 50 samples from the features
np.random.seed(42)  # For reproducibility
random_indices = np.random.choice(features.shape[0], size=50, replace=False)
features_subset = features[random_indices]

# Standardize the subset of features
scaler = StandardScaler()
features_scaled_subset = scaler.fit_transform(features_subset)

# Perform Agglomerative Clustering on the subset
agglom = AgglomerativeClustering(n_clusters=None, affinity='euclidean', linkage='ward', distance_threshold=0)
agglom.fit(features_scaled_subset)

# Create linkage matrix for the dendrogram using the subset
linked_subset = linkage(features_scaled_subset, 'ward')

# Plot the dendrogram for the subset
plt.figure(figsize=(10, 7))
dendrogram(linked_subset, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram (Subset of 50)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
import csv
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Path to your CSV file
file_path = '/content/drive/Shareddrives/DATA 240- Data Mining/Cleaned dataset and code/Canada /Final Cleaned/full_cleaned_normalized4.csv'  # Update with your actual file path

# Read the data from the CSV file
with open(file_path, mode='r') as file:
    reader = csv.DictReader(file)
    data = list(reader)

# Extract 'sentiment_y' and 'normalized_popularity' for clustering
sentiment_ys = [float(item['sentiment_y']) for item in data]
normalized_popularities = [float(item['normalized_popularity']) for item in data]

# Combine sentiment_ys and normalized_popularities into a single NumPy array
features = np.array(list(zip(normalized_popularities, sentiment_ys)))

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

k=4

# Perform K-Means Clustering on the scaled features
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(features_scaled)

# The cluster labels are available via kmeans.labels_
cluster_labels = kmeans.labels_

# Plot the clusters along with the centroids
plt.figure(figsize=(10, 7))

# Create a scatter plot for each cluster
for i in range(k):
    # Select only data points that belong to the current cluster
    cluster_data = features_scaled[cluster_labels == i]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {i}')

# Plot centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='black', marker='X', label='Centroids')

# Plot formatting
plt.title('K-Means Clustering (Sentiment_y vs Normalized Popularity)')
plt.xlabel('Normalized Popularity (scaled)')
plt.ylabel('Sentiment Score (scaled)')
plt.legend()
plt.show()


silhouette_avg = silhouette_score(features_scaled, cluster_labels)

print(f"The average silhouette score for k={k} is: {silhouette_avg}")
```
